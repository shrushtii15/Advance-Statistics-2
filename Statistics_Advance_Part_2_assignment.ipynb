**Statistics Part 2**

1. What is hypothesis testing in statistics?

  Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  
   Null Hypothesis (H0):Represents no effect, no difference, or status quo

  Alternative Hypothesis (H1 or Ha)Represents the effect, difference, or claim being tested



3. What is the significance level in hypothesis testing, and why is it important?

  Denoted as α (commonly 0.05 or 5%).

  Represents the probability of making a Type I error (rejecting a true null hypothesis).

4. What does a P-value represent in hypothesis testing?

  The P-value is the probability of observing your test results (or more extreme) if the null hypothesis H0 is true.



5. How do you interpret the P-value in hypothesis testing?

 P≤α:Reject 𝐻0;evidence supports 𝐻1

 P>α: Fail to reject H0; not enough evidence for H1
​



6. What are Type 1 and Type 2 errors in hypothesis testing?

 Type I-error : Rejecting true H0.

 Tpye II-Error: Failing to reject a false H0


7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

   One-Tailed: Tests effect in one direction.

   Two-tailed: Tests in both direction

8. What is the Z-test, and when is it used in hypothesis testing?

  A Z-test checks if the sample mean significantly differs from the population mean when population standard deviation (σ) is known and sample size is large (n ≥ 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

  Z= (X−μ)/σ/root(n)

  Represents: How many standard errors the sample mean is from the population mean
​


10. What is the T-distribution, and when should it be used instead of the normal distribution?

  Use T-distribution when the population standard deviation (σ) is unknown and sample size is small (n < 30).

11. What is the difference between a Z-test and a T-test?

 Z-test:known population SD , Large sample sizes.

 T-test: Unknown population SD, Small Sample sizes



12. What is the T-test, and how is it used in hypothesis testing?

  A T-test compares sample means to a known population mean or between two groups to see if they are significantly different.

  Types:One-sample t-test

  Two-sample t-test (independent)

  Paired t-test (related groups)

13. What is the relationship between Z-test and T-test in hypothesis testing?

  Both test hypotheses about means.
  T-test is used when variance is unknown and/or sample is small.
  As sample size increases, t-distribution approximates normal (Z).



14. What is a confidence interval, and how is it used to interpret statistical results?

  A confidence interval (CI) provides a range of values for an unknown population parameter, with a given level of confidence (e.g., 95%).

  Example: "We are 95% confident the population mean lies between 45 and 55."

15. What is the margin of error, and how does it affect the confidence interval?

  The margin of error (ME) is the range added/subtracted to the sample estimate to create the CI.

  Larger margin → wider CI, showing more uncertainty.

16. How is Bayes' Theorem used in statistics, and what is its significance?

  Bayes' Theorem updates the probability of a hypothesis based on new evidence.

  Applications: Medical diagnosis, spam filtering, decision making under uncertainty.

17. What is the Chi-square distribution, and when is it used?

  A right-skewed distribution used in tests of categorical data.

  Applications:

  Test for variance

  Goodness-of-fit

  Test of independence in contingency tables.



18. What is the Chi-square goodness of fit test, and how is it applied?

  It checks if observed frequencies match expected frequencies for a categorical variable.
  
  difference  of observered value and Expected value By expected value

19. What is the F-distribution, and when is it used in hypothesis testing?

  The F-distribution is right-skewed and used when comparing variances between two or more groups.

  Applications:

  ANOVA (Analysis of Variance)

  F-test for variance comparison

20. What is an ANOVA test, and what are its assumptions?

  ANOVA (Analysis of Variance) compares means across multiple groups.

  Assumptions:

  Independence of observations

  Normally distributed groups

  Equal variances (homogeneity of variance)

21. What are the different types of ANOVA tests?

  One-way ANOVA: One independent variable with multiple levels (groups).

  Two-way ANOVA: Two independent variables.

  Repeated measures ANOVA: Same subjects tested across conditions.

22. What is the F-test, and how does it relate to hypothesis testing?

  The F-test compares two variances or group means in ANOVA.
  It tests whether the variances or group means are significantly different.

```python
#1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results

from scipy.stats import norm
import numpy as np

sample = np.random.normal(70, 10, 30)
pop_mean = 68
pop_std = 10

sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
#2.  Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
data = np.random.normal(100, 15, 50)
pop_mean = 95
z_stat = (np.mean(data) - pop_mean) / (15 / np.sqrt(len(data)))
p_val = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_val}")

```

```python
#3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean
def one_sample_z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    n = len(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

sample = np.random.normal(70, 10, 40)
z, p = one_sample_z_test(sample, 68, 10)
print(f"Z-score: {z}, P-value: {p}")

```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Curve')
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

plt.axvline(z_critical, color='red', linestyle='dashed', label=f'Z-critical (+{z_critical:.2f})')
plt.axvline(-z_critical, color='red', linestyle='dashed', label=f'Z-critical (-{z_critical:.2f})')
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3)
plt.title('Two-tailed Z-Test Decision Region')
plt.legend()
plt.show()

```

```python
#5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    y_alt = norm.pdf(x, 1.5, 1)  # Alternate distribution shifted right

    plt.plot(x, y, label='Null Hypothesis (H0)')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H1)')

    alpha = 0.05
    z_critical = norm.ppf(1 - alpha)

    plt.axvline(z_critical, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y_alt, where=(x < z_critical), color='blue', alpha=0.3, label='Type II Error')

    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.show()

visualize_errors()

```

```python
#9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return ci

data = np.random.normal(100, 15, 30)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
#10.  Write a Python program to calculate the margin of error for a given confidence level using sample data
sample = np.random.normal(100, 10, 40)
sem = stats.sem(sample)
confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=len(sample)-1)

margin_error = t_critical * sem
print(f"Margin of Error: {margin_error}")
```

```python
#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + likelihood_B_given_not_A * prior_not_A
    return numerator / denominator

# Example:
P_A = 0.01         # Disease prevalence
P_B_given_A = 0.99  # Test sensitivity
P_not_A = 0.99
P_B_given_not_A = 0.05  # False positive rate

posterior = bayes_theorem(P_A, P_B_given_A, P_not_A, P_B_given_not_A)
print(f"Posterior Probability of having disease given a positive test: {posterior}")
```
